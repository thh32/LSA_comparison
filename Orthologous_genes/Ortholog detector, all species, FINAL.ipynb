```python
"""
Re-checked method
"""

files = ''
if animal == 'chimp':
    files = 'Ortholog_output/Human_to_chimp_nonfilt_orthologs.tsv'
elif animal == 'cow':
    files = 'Ortholog_output/Human_to_cow_nonfilt_orthologs.tsv'
elif animal == 'chicken':
    files = 'Ortholog_output/Human_to_chicken_nonfilt_orthologs.tsv'
elif animal == 'platypus':
    files = 'Ortholog_output/Human_to_platypus_nonfilt_orthologs.tsv'
elif animal == 'zebrafish':
    files = 'Ortholog_output/Human_to_zebrafish_nonfilt_orthologs.tsv'
elif animal == 'wallaby':
    files = 'Ortholog_output/Human_to_wallaby_nonfilt_orthologs.tsv'
    
    

# Create list to find 121 orthologs in orthlog list
counts = {}
count =0
for line in open(files,'r'):
    count +=1
    if count == 1:
        continue
    else:
        timber = line.strip('\n').split('\t')
        for i in timber:
            if i in counts:
                counts[i] +=1
            else:
                counts[i] = 1
                


# Create checked list of 121 orthlogs which are correct
count = 0
true = 0
chimp_true = []


    
    
print files
for line in open(files,'r'):
    count +=1
    if count == 1:
        continue
    else:
        timber = line.strip('\n').split('\t')
        if counts[timber[0]] == 1: # This ensures that only 121 orthologs are selected
            if counts[timber[1]] == 1:
                true +=1
                chimp_true.append(timber[1])

                
# Compare found orthologs to known orthlogs
ortho_count = 0
for i in found:
    if i in chimp_true:
        ortho_count +=1
        
print "True positive =  ", ortho_count

TP = ortho_count
FP = len(file2_tophits2) - ortho_count
FN = len(chimp_true) - ortho_count

print "False positive = ", FP
print "False negative = ", FN
print "Speficicity = ", (TP/(TP+FP)) 
print "Sensitivity = ", (TP/(TP+FN)) 
print "Relative error Quotient = ", (FN+FP)/(TP*2)
print 'Recall = ', (TP/(TP+FN))
print 'Precision = ', (TP/(TP+FP))

precision = (TP/(TP+FP))
recall = (TP/(TP+FN))

Fmeasure =  2*((precision*recall)/precision+recall) 


print 'F-measure = ', Fmeasure

```
